```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
import pickle
```

```python
data=pd.read_csv(r"C:\Users\Pushkar S\OneDrive\Desktop\Data science Practice\Project\Churn_Modelling.csv")
data.head(5)
```

```python
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
data
```

```python
label_encoder_gemder=LabelEncoder()
data["Gender"]=label_encoder_gemder.fit_transform(data['Gender'])
data
```

```python
data
```

```python
onehot_encoder_geo=OneHotEncoder()
geo_encoder=onehot_encoder_geo.fit_transform(data[['Geography']])
geo_encoder
```

```python
onehot_encoder_geo.get_feature_names_out(['Geography'])
```

```python
geo_encode=pd.DataFrame(geo_encoder.toarray(),columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encode
```

```python
data=pd.concat([data.drop('Geography',axis=1),geo_encode],axis=1)
data
```

```python
with open('label_encoder_gemder.pkl','wb') as file:
    pickle.dump(label_encoder_gemder,file)

with open('onehot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_encoder_geo,file)
```

```python
data.head()
```

```python
x=data.drop('Exited',axis=1)
y=data['Exited']
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
stdscalar=StandardScaler()
x_train=stdscalar.fit_transform(x_train)
x_test=stdscalar.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
log_dir="logs/fit"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_stopping_callback=EarlyStopping(monitor='val_loss',patience=15,restore_best_weights=True)
```

```python
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping_callback]
)
```
